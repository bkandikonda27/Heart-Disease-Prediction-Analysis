```python
# the min value for RestingBP and Cholesterol should not be 0, so I will replace these values with the median or
# remove it compelted from the data set if it is only one value
fix_restingBP = df[df['RestingBP'] == 0].value_counts()
df.drop(df.loc[df['RestingBP']==0].index, inplace=True)
df.describe()
```

```python
chol_median = df[df.Cholesterol != 0]
chol_median['Cholesterol'].median()
```

```python
df['Cholesterol'].replace(0, chol_median['Cholesterol'].median(), inplace=True)
df.describe()
```

Here, I will split the quantitative features I want to dig into, into quartiles. So, I can dig into each feature
based on an individuals quartile range within the data set.

```python
df['RestingBPQuartile'] = pd.qcut(df['RestingBP'], 4, ['1Q','2Q','3Q','4Q'])
df['CholesterolQuartile'] = pd.qcut(df['Cholesterol'], 4, ['1Q','2Q','3Q','4Q'])
df['MaxHRQuartile'] = pd.qcut(df['MaxHR'], 4, ['1Q','2Q','3Q','4Q'])
df['AgeQuartile'] = pd.qcut(df['Age'], 4, ['1Q','2Q','3Q','4Q'])
```
