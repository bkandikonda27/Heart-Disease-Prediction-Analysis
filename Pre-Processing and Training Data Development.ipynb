```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

print('loaded')
```

```python
df = pd.read_csv('data/HeartDataWrangling.csv', index_col = 0)
df.head()
```

```python
df.info()
```

1. Creating Dummy Variables for my Categorical features

```python
cat_features = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']
new_df = pd.get_dummies(df, columns = cat_features)
new_df.dtypes
```

I am dropping the following columns for data integrity purposes. For example, if 'Sex_F' is 0 we already know that
the person is a 'Male' so if we keep 'Sex_M', we are presenting the same data twice. It is the same for all of the
other categorical variables mentioned above. I am dropping one column in each of the categorical variables.

```python
new_df.drop(['Sex_M', 'ChestPainType_TA', 'RestingECG_ST', 'ExerciseAngina_N', 'ST_Slope_Down'], axis=1, inplace=True)
```

```python
new_df.head()
```

I am going to drop all of the 'Quartile' columns I added through qcut in the EDA section. This is so there are no 'string' columns in the data set.

```python
new_df.drop(['RestingBPQuartile', 'CholesterolQuartile', 'MaxHRQuartile', 'AgeQuartile'], axis='columns', inplace=True)
new_df.head()
```

3. Standardizing the Magnitude of Numeric Features

```python
num_col = ['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak']
for col in num_col:
    df[col].plot(kind = 'hist', title = col)
    plt.show()
```

I want to standardize age, restingbp, cholesterol, maxhr, and oldpeak. I am going to use min-max scaler to do this because not all of the data falls under normal distribution

```python
mms = MinMaxScaler()

def scaler(scaler, df, col_name, new_col_name):
    scaler.fit(df[[col_name]])
    df[new_col_name] = scaler.transform(df[[col_name]])

print('done')
```

```python
scaler(mms, new_df, 'Age', 'scaled_Age')
scaler(mms, new_df, 'RestingBP', 'scaled_RestingBP')
scaler(mms, new_df, 'Cholesterol', 'scaled_Cholesterol')
scaler(mms, new_df, 'MaxHR', 'scaled_MaxHR')
scaler(mms, new_df, 'Oldpeak', 'scaled_Oldpeak')

new_df.head()
```

```python
new_df = new_df.drop(['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak'], axis=1)
new_df.head()
```

My final step in the pre-processing module is to create a training and testing split

```python
y = new_df[['HeartDisease']]
X = new_df.drop('HeartDisease', axis=1)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 17, test_size = 0.3)
```
